#### ............................................................ Get countries only............................................................ 
- Remove top pages

```python
c_index = []
def countryonly(list):
    for i in country_names:
        try:
            t = re.search(i, text)
            c_index.append(t.span()[0]) #append the starting positions of the country match
        except:
            pass
    print(c_index)
```

```python
countryonly(country_names)   #we will know US index is 73409, CAN index is 63109
```

```python
all_countries = text[min(c_index):] #print CAN & US text in text
```
