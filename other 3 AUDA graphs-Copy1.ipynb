```python
import seaborn as sns
from Pardee_Package import color_palletes as pardee_colors
from Pardee_Package import style_dictionaries as pardee_styles
import matplotlib.pyplot as plt
```

```python
def high_contrast_color_pallete(): 
    red = "#cc2735"
    blue = "#099bd7"
    green = "#3ea248"  
    yellow = "#f0e14a"
    orange = "#f78b29"
    purple = "#61459c"
    #gray = "#898989"
    #light_blue = "#A2C8EC"
    #light_orange = "#FFB579"
    #light_gray = "#CFCFCF"
    color_pallete = [red,
                     blue,
                     green,
                     yellow,
                     orange,
                     purple]
                     #gray
                     #gray,
                     #light_blue,
                     #light_orange,
                     #light_gray]
    return color_pallete

def off_white():
    return "#F4F3ED"

standard_style = {
'axes.facecolor': off_white(),
'axes.edgecolor': '.8',
'axes.grid': True,
'axes.axisbelow': True,
'axes.labelcolor': '.15',
'figure.facecolor': 'white',
'grid.color': '.8',
'grid.linestyle': 'solid',
'text.color': '.15',
'xtick.color': '.15',
'ytick.color': '.15',
'xtick.direction': 'out',
'ytick.direction': 'out',
'lines.solid_capstyle': 'round',
'patch.edgecolor': 'w',
'patch.force_edgecolor': True,
'image.cmap': 'rocket',
'font.family': ['Trebuchet MS'],
'font.Tebuchet MS': ['Trebuchet MS'],
'font.scale': '9',
#'font.size': 19,
'xtick.bottom': False,
'xtick.top': False,
'ytick.left': False,
'ytick.right': False,
'axes.spines.left': True,
'axes.spines.bottom': True,
'axes.spines.right': True,
'axes.spines.top': True
    }

 


#Example Code for setting seaborn pallet and style:

sns.set_style(standard_style)
```

```python
sns.set(rc={'figure.figsize':(15,8)})
sns.set(font_scale = 1.4)
sns.set_style(standard_style)
sns.set_palette(high_contrast_color_pallete())
ax = sns.lineplot(x="Year", y="value", hue="variable", style="variable",
                       palette=high_contrast_color_pallete(),
                      data=df2)
handles, labels = ax.get_legend_handles_labels()
ax.legend(handles=handles[1:],labels=labels[1:], title="Scenario")
ax.set_ylabel("GDP growth rate (MER $2011)")
ax.set(xlabel=None)
ax.set_title('GDP Growth by Year and COVID Scenario')
figure = ax.get_figure()    
figure.savefig('figure 6.png', dpi=300)
plt.show()
```
